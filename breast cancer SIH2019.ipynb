```python
## variables that i will be using for the model after checking correlation, multicollinearity and relevancy check
prediction_var = ['texture_mean','perimeter_mean','smoothness_mean','compactness_mean','symmetry_mean']
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(can_data, test_size = 0.3,random_state=101)
```

```python
### understand training and test dataset
print(train.shape)
print(test.shape)
```

```python
test.columns
```

```python
train_x = train[prediction_var]
train_y = train.diagnosis
test_x  = test[prediction_var]
test_y  = test.diagnosis
```

```python
### perform logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
logreg_mod = LogisticRegression()
```

```python
### fit the model
logreg_mod.fit(train_x,train_y)
```

```python
x = can_data.iloc[:, 1:31].values
y = can_data.iloc[:, 0].values
y
```

```python
### run predictions on test
logreg_mod_predictions = logreg_mod.predict(test_x)
```

```python
### evaluate the model
from sklearn import metrics
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
### output the confusion matrix
###  confusion matrix is as below

###    true negative               false positve
####   false negaitve              true  positive
confusion_matrix(test_y,logreg_mod_predictions)

### out of confusion matrix run 
### 101 observations are correctly predicted as "0" benign - TRUE NEGATIVE
###  4  observations are wrongly   predicted as "1" Malign  - False Positive
###  15 observations are wrong     predicted as  "0" Benigh - False Negative
### 51  observations are correctly predicted as "1" Malign  - True positive.

### accuracy is calculted as  True Negative + True Positive)/total observations          = (101+51 )/ (101+51+4+15) = 152/171=88.8%
### error is                  False Negative + False Positive)/ Total Observations       = 15+4/171 =11%
```

```python
logreg_mod.score(test_x,test_y)
```

```python
(101+51 )/(101+51+4+15)
```

```python
print(classification_report(test_y,logreg_mod_predictions))

### Out put of classification Report - Interpretation
### FROM the confusion matrix we are able to deduce - 116 (101+15) observations belong to class 0 - Benign
####                                                   61 (4 + 51) observations belong to class 1- Malign

#### The precision is intuitively the ability of the classifier not to label as positive a sample that is negative
###  precision = tp / (tp + fp)
###  precision for class 1 Malign =   51/(51(TP)+4(FP)) = .93
### precision for class 0 Benign =    101/(101(TN)+15(FN)) = .87

#### The recall is intuitively the ability of the classifier to find all the positive samples.
### RECALL for class 1 =  True Positive/ (True Positive + False Negative) = 51/(51+15) = .77
###  RECALL for class 0 = True Negative /True Negative + False Positive) = 101/(101+4) = .96

#### The F-beta score can be interpreted as a weighted harmonic mean of the precision and recall, where an F-beta score reaches its best value at 1 and worst score at 0.

#### The F-beta score weights recall more than precision by a factor of beta. beta == 1.0 means recall and precision are equally important.

### The support is the number of occurrences of each class in y_true.
```
